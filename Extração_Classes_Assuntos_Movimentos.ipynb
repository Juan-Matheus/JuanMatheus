Crio um função que vai receber um parâmetro, assim é reponsável por extrair as listas de classes, assuntos ou movimentos do site

```python
def extrair_listas_jus(tipo_tabela):
    url = f"https://www.cnj.jus.br/sgt/versoes.php?tipo_tabela={tipo_tabela}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')



    listas = []
    tabelas = soup.find_all('table')

    for tabela in tabelas:
        linhas = tabela.find_all('tr')
        for linha in linhas:
            colunas = linha.find_all('td')
            lista = []
            for coluna in colunas:
                lista.append(coluna.text.strip())
            if lista:
                listas.append(" | ".join(lista))

    return listas
```
