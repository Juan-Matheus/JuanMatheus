```python
import numpy as np
import pandas as pd
```

Explorando os dados

```python
df = pd.read_csv('/agricultural_raw_material.csv')
```

```python
# Contagem dos dados nulos
df.info
df.isnull().sum()
```

Como lidar com dados incorretos, inv√°lidos e ausentes.

```python
# REPLACE
df = df.replace('%','', regex=True)
df = df.replace(',','', regex=True)
df = df.replace('-','', regex=True)
df = df.replace('', np.nan)
# descartando linhas com valores NaN
df = df.dropna()
# verifique se todos os valores Nan foram resolvidos
df.isnull().sum()
# Convertendo tipo de dado para float
list = ["Coarse wool Price", 
"Coarse wool price % Change", 
"Copra Price", 
"Copra price % Change",
"Cotton price % Change",
"Fine wool Price", 
"Fine wool price % Change", 
"Hard log price % Change",
"Hard sawnwood price % Change",
"Hide price % change",
"Plywood price % Change",
"Rubber price % Change",
"Softlog price % Change",
"Soft sawnwood price % Change",
"Wood pulp price % Change"]
df[list] = df[list].astype("float")
df.dtypes
```

```python
df.head()
```

COLUNAS DE DATA E HORA 
Formatando a coluna datetime e definindo com √≠ndice para conjunto de dados

```python
df.Month = pd.to_datetime(df.Month.apply(str).str.upper(), format= '%b%y', yearfirst=False)

# Indexing Month

df = df.set_index('Month')
```

```python
df.head()
```

An√°lise Explorat√≥ria e Visualiza√ß√£o

```python
from ast import increment_lineno
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

MAPA DE CALOR

```python
#raw-materials list
raw_data=['Coarse wool Price',  'Copra Price','Cotton Price', 'Fine wool Price',  'Hard log Price', 'Hard sawnwood Price',
        'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']

# Obtendo a matriz de correla√ß√£o
corrmat = df[raw_data].corr()

# setting the size of plot
fig = plt.figure(figsize = (12,9))

#maskinf the uppwe traingle part since matrix is symmetric(repetitive)
mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax = .8, mask=mask,square = True, annot = True)
plt.show()
```

O mapa de calor mostra a correla√ß√£o entre as mat√©rias-primas maior o valor correlacionado maior chance de serem duas mat√©rias-primas relacionadas.

```python
plt.figure(figsize=(30,15))
changelist=['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change','Fine wool price % Change',
            'Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change']

# Generate a correlation matrix for the whole dataset
corrMatrix = df[changelist].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

A partir deste gr√°fico podemos dizer que quase n√£o h√° rela√ß√£o entre a varia√ß√£o percentual dos pre√ßos das mat√©rias-primas.

1.   o valor negativo implica que duas vari√°veis ‚Äã‚Äãest√£o negativamente correlacionadas (um aumento, outro decr√©scimo).
2.   Zero implica nenhuma rela√ß√£o.
3.   caso contr√°rio, maior o valor maior a chance de rela√ß√£o, pre√ßos e seus gr√°ficos de % de mudan√ßa.

Mat√©ria-Prima que ten o menor pre√ßo ao longo dos anos

```python
plt.figure(figsize=(15,10))
materialslist = ['Coarse wool Price',  'Copra Price','Cotton Price', 'Fine wool Price',  'Hard log Price', 'Hard sawnwood Price',
        'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']
for i in range(len(materialslist)):
  plt.subplot(4,3,i+1)
  plt.subplots_adjust(hspace=1,wspace = 0.5)
  plt.title(materialslist[i])
  plt.plot(df[materialslist[i]])
  plt.xticks(rotation = 90)
  plt.suptitle("Compara√ß√£o de menor pre√ßo por mat√©ria prima ao longo dos anos")
```

O algod√£o e a borracha s√£o os pre√ßos mais baixos, vamos comparar os pre√ßos para identificar o menor entre eles.

```python
plt.figure(figsize=(10,10))
plt.plot(df[['Cotton Price', 'Rubber Price']])
plt.title("Compara√ß√£o de Pre√ßo Cotton x Rubber")
plt.xlabel('Years')
plt.ylabel('Prices')
plt.legend(['Cotton Price','Rubber Price'],loc='upper center',bbox_to_anchor=(1.2,1))
```
